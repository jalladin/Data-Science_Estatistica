**Correlacion**

- Correlacion positiva:
    - Quando duas variveis se movem na mesma direçao.
    - QUando o valor de uma variavel aumenta, o valor da outra tambem aumenta.
-  Correlacion negativa:
    - As variaveis se movem em em direçao oposta.
    - Quando o valor de uma variavel aumenta o outro diminui
- Correlaçao positiva perfeita
    - As proporcoes sao identicas e na mesma direçao.
- Correlaçao negativa perfeita
    - As proporçoes sao identicas mas a direçao é oposta.
- Correlaçcao Nula
    - Quando nao existe relacao ou é muito debil

**Coeficiente da correlaçao:**

- r pode oscilar entre -1 e 1.
- Quanto mais perto de -1 maior a correlacao negativa
- Quanto mais perto de 1 maior a correlacao positiva
- Quanto mais perto de 0 menos correlaçao
- Se r = 0, nao existe correlaçao entre as variaveis
- Se -0,75 >=r >= -1 existe uma correlacao negativa alta
- Se 0,75 <= r <= 1 existe uma correlacao positiva alta
